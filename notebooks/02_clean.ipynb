```python
# Clears data frame removing rows with null values
df_clean = df.dropna()
df_clean.isnull().sum()
```

```python
# Convert 'ano' column to int
df_clean = df_clean.assign(ano = df_clean['ano'].astype(int))

import pycountry

# Function to convert ISO2 country codes to ISO3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

df_clean['residencia_iso3'] = df_clean['residencia'].apply(iso2_to_iso3)
df_clean.to_csv(
'../data/processed/data_processed.csv',
index=False
)

df_clean.head()
```
