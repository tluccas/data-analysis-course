```python
lines, columns = df.shape[0], df.shape[1]

print(f"Lines: {lines}\nColumns: {columns}")
```

```python
translate_columns = {
    'work_year': 'ano',
'experience_level': 'senioridade',
'employment_type': 'contrato',
'job_title': 'cargo',
'salary': 'salario',
'salary_currency': 'moeda',
'salary_in_usd': 'salario_usd',
'employee_residence': 'residencia',
'remote_ratio': 'remoto',
'company_location': 'empresa',
'company_size': 'tamanho_empresa',
}

df = df.rename(columns=translate_columns)
print("Colunas:\n")
for i, column in enumerate(df.columns):
    print(f"{i + 1}. {column}")
```

```python
df["senioridade"].value_counts()
```

```python
translate_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df["contrato"] = df["contrato"].replace(translate_contrato)
df["contrato"].value_counts()
```

```python
translate_tamanho_empresa = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}
df["tamanho_empresa"] = df["tamanho_empresa"].replace(translate_tamanho_empresa)
df["tamanho_empresa"].value_counts()
```

```python
translate_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df["senioridade"] = df["senioridade"].replace(translate_senioridade)
df["senioridade"].value_counts()
```

```python
translate_modelo_trabalho = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}
df["remoto"] = df["remoto"].replace(translate_modelo_trabalho)
df["remoto"].value_counts()
```
